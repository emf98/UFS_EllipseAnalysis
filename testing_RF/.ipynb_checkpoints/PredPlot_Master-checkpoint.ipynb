```python
##import files to make the plots
##data management
from PredictionPlots import extrapolate, daily_anomaly, preprocess_andopen
##plotting
from PredictionPlots import  BWcheckevent_label, CScheckevent_label, BWplot, GPH_horzCS
```

```python
##for horizontal cross sections of GPH
infile = open("../UFS_metrics/CAP_ERA5_GPH10.p","rb")
gph_input = pickle.load(infile)  ##GPH vertical cross section along longitudes
GPH = gph_input[:, :, 0:18, :, :].reshape((56*18, 26, 180)) 
infile.close()

infile = open("../UFS_metrics/CAP_ERA5_GPHanoms10.p","rb")
gph_input = pickle.load(infile)  ##GPH vertical cross section along longitudes
GPHA = gph_input[:, :,0:18, :, :].reshape((56*18, 26, 180)) 
infile.close()

print("GPH shape: ",GPH.shape)
```

```python
lat = np.arange(90, 38, -2)
print("Len (90-40) Latitudes: ", len(lat))
lon = np.arange(0, 362, 2)
print("Len longitudes: ", len(lon))
#lev = np.array([1., 2., 3., 5., 7., 10., 20., 30., 50., 70., 100., 125., 150., 175., 200., 225., 250., 300., 350., 400., 
                #450., 500., 550., 600., 650., 700., 750., 775., 800., 825., 850., 875., 900., 925., 950., 975., 1000.])
#print("Len levels: ", len(lev))
```

```python
files = ['ERA5','Prototype5','Prototype6','Prototype7','Prototype8']
region = ['Europe','Canada','SEUS']
```

```python
for i in range(len(region)):
    for j in range(len(files)):
        forecast_area = region[i] #should be Europe, Canada, or SEUS
        region_file = str(f'{forecast_area}_{files[j]}') ##ex. Canada_ERA5
        
        if files[j] == str('ERA5'):
            ufs_index = " "
            input_data, output_data, metrics, pos, Fpos, neg, Fneg = preprocess_andopen(forecast_area, str("ERA5"), ufs_index, region_file)
            #classify events
            print("Separating indices ...")
            BW_pos = BWcheckevent_label(pos,input_data)
            BW_neg = BWcheckevent_label(neg,input_data)
            BW_Fpos = BWcheckevent_label(Fpos,input_data)
            BW_Fneg = BWcheckevent_label(Fneg,input_data)
            #BW plot
            print("Plotting ...")
            BWplot(BW_pos, BW_neg, BW_Fpos, BW_Fneg, metrics,
                   f' {region[i]} {files[j]}',f'./images/composite/BW/BW_{region_file}')

            ##GPH Cross Section
            #horizontal
            GPH_cpos = CScheckevent_label(pos,GPH)
            GPH_cneg = CScheckevent_label(neg,GPH)
            GPH_Fpos = CScheckevent_label(Fpos,GPH)
            GPH_Fneg = CScheckevent_label(Fneg,GPH)
            
            GPHA_cpos = CScheckevent_label(pos,GPHA)
            GPHA_cneg = CScheckevent_label(neg,GPHA)
            GPHA_Fpos = CScheckevent_label(Fpos,GPHA)
            GPHA_Fneg = CScheckevent_label(Fneg,GPHA)
            
            GPH_horzCS(GPH_cpos,GPH_cneg,GPH_Fpos,GPH_Fneg,GPHA_cpos,GPHA_cneg,GPHA_Fpos,GPHA_Fneg,
                           -1000,1000,100,f' {region[i]} {files[j]}', lat, lon,f'./images/composite/gph/horzGPH_{region_file}')

        else:
            ufs_index = j-1 ##add index to indicate which UFS Prototype is being observed.
            input_data, output_data, metrics, pos, Fpos, neg, Fneg = preprocess_andopen(forecast_area, str("UFS"), ufs_index, region_file)
            #classify events
            print("Separating indices ...")
            BW_pos = BWcheckevent_label(pos,input_data)
            BW_neg = BWcheckevent_label(neg,input_data)
            BW_Fpos = BWcheckevent_label(Fpos,input_data)
            BW_Fneg = BWcheckevent_label(Fneg,input_data)
            #BW plot
            print("Plotting ...")
            BWplot(BW_pos, BW_neg, BW_Fpos, BW_Fneg, metrics,
                   f' {region[i]} {files[j]}',f'./images/composite/BW/BW_{region_file}')

            ##GPH Cross Section
            #horizontal
            GPH_cpos = CScheckevent_label(pos,GPH)
            GPH_cneg = CScheckevent_label(neg,GPH)
            GPH_Fpos = CScheckevent_label(Fpos,GPH)
            GPH_Fneg = CScheckevent_label(Fneg,GPH)
            
            GPHA_cpos = CScheckevent_label(pos,GPHA)
            GPHA_cneg = CScheckevent_label(neg,GPHA)
            GPHA_Fpos = CScheckevent_label(Fpos,GPHA)
            GPHA_Fneg = CScheckevent_label(Fneg,GPHA)
            
            GPH_horzCS(GPH_cpos,GPH_cneg,GPH_Fpos,GPH_Fneg,GPHA_cpos,GPHA_cneg,GPHA_Fpos,GPHA_Fneg,
                           -1000,1000,100,f' {region[i]} {files[j]}', lat, lon,f'./images/composite/gph/horzGPH_{region_file}')
```
